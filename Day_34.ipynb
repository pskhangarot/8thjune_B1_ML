```python
import nltk
import pyttsx3
import speech_recognition as sr
import warnings
warnings.filterwarnings('ignore')
```

```python
from nltk.chat.util import Chat,reflections
```

```python
reflections
```

```python
pairs=(
    (r'i like(.*)',
        (
            'Surely you can like %1'
            'why, do you like %1?'
        )
    ),
    (r'will you please(.*)',
        (
            'yeah,sure',
            'no,i can not do this'
        )
    ),
    (r'(.*)python(.*)',
        (
            'python is an interpreted language',
            'python is 4th generation language',
            'python is language of totday and tomorrow'
        )
    )
)
```

```python
cb=Chat(pairs,reflections)
cb
```

```python
cb.converse(quit='bye')
```

```python
cb.respond('tell me about python')
```

```python
speng=pyttsx3.init()
```

```python
speng.setProperty('rate',120)
```

```python
qu='can python be used for machine learning'
```

```python
resp=cb.respond(qu)
speng.say(resp)
speng.runAndWait()
resp
```

```python
speng.say('hello sir, what can i do for you')
speng.runAndWait()
```

```python
while True:
    with sr.Microphone() as source:
        r.adjust_for_ambient_noise(source)
        audio=r.listen(source)
        try:
            que=r.recognize_google(audio)
            if que=='bye':
                break
            print('you said:',que)
            resp=cb.respond(que)
            if resp:
                print('Bot:',resp)
                speng.say(resp)
                speng.runAndWait()
            else:
                if 'search' in que:
                    if 'chatbot' in que:
                        wb.open_new('http://google.com/search?q=chatbot')
                    elif 'games' in que:
                        wb.open_new('http://google.com/search?q=chatbot')

                elif'purchase' in que:
                    query=que.split('purchase')[-1].strip()
                    wb.open_new('http://flipkart.com/search?q={}'.format(query))
            print('-'*20)
        except Exception as e:
            print('error')
            print(e)
```
