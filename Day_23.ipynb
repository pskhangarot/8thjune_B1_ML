```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
x = 2*np.random.normal(0,1,30)
x
```

```python
plt.plot(x)
```

```python
y = x-3*(x**2) + 0.6*(x**3) + np.random.normal(0,1,30)
y
```

```python
plt.scatter(x,y,s=25,c='r')
```

```python
x=x[:,np.newaxis]
x.shape
```

```python
y=y[:,np.newaxis]
y.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model1=LinearRegression().fit(x,y)
```

```python
ypred = model1.predict(x)
```

```python
model1.score(x,y)*100
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y,s=25,c='r')
plt.plot(x,ypred,c='g')
plt.grid()
```

```python
from sklearn import preprocessing as pp
```

```python
x.shape
```

```python
x
```

```python
x_poly1=pp.PolynomialFeatures(1).fit_transform(x)
x_poly1
```

```python
x_poly1=pp.PolynomialFeatures(1).fit_transform(x)
model_d1=LinearRegression().fit(x_poly1,y)
ypred_d1=model_d1.predict(x_poly1)
print(100*model_d1.score(x_poly1,y))
plt.figure(figsize=(10,6))
plt.scatter(x,y,s=25,c='r')
plt.plot(x,ypred_d1,c='g')
plt.grid()
```

```python
x_poly2=pp.PolynomialFeatures(2).fit_transform(x)
model_d2=LinearRegression().fit(x_poly2,y)
ypred_d2=model_d2.predict(x_poly2)
print(100*model_d2.score(x_poly2,y))
plt.figure(figsize=(10,6))
plt.scatter(x,y,s=25,c='r')
plt.plot(x,ypred_d1,c='b')
plt.scatter(x,ypred_d2,c='g')
plt.grid()
```

```python
x_poly3=pp.PolynomialFeatures(3).fit_transform(x)
model_d3=LinearRegression().fit(x_poly3,y)
ypred_d3=model_d3.predict(x_poly3)
print(100*model_d3.score(x_poly3,y))
plt.figure(figsize=(10,6))
plt.scatter(x,y,s=25,c='r')
#plt.plot(x,ypred_d1,c='b')
plt.scatter(x,ypred_d2,c='g')
plt.grid()
```

```python
x_poly4=pp.PolynomialFeatures(4).fit_transform(x)
model_d4=LinearRegression().fit(x_poly4,y)
ypred_d4=model_d4.predict(x_poly4)
print(100*model_d4.score(x_poly4,y))
plt.figure(figsize=(10,6))
plt.scatter(x,y,s=25,c='r')
#plt.plot(x,ypred_d1,c='b')
plt.scatter(x,ypred_d2,c='g')
plt.grid()
```

```python
score = []
for i in range(1,10):
    x_poly_i = pp.PolynomialFeatures(i).fit_transform(x)
    model_d_i = LinearRegression().fit(x_poly_i,y)
    score.append(100*model_d_i.score(x_poly_i,y))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,10),score,marker='*',)
plt.grid()
```

```python
df['State'].str.get_dummies()
```

```python
df_proc=pd.concat([df,df['State'].str.get_dummies()],axis=1)
df_proc.head()
```

```python
df_proc.drop('State',axis=1,inplace=True)
df_proc.head()
```

```python
x=df_proc.drop('Profit',axis=1)
y=df_proc['Profit']
```

```python
model=LinearRegression().fit(x,y)
```

```python
model.score(x,y)*100
```

```python
rnd=pd.DataFrame(pp.PolynomialFeatures(3).fit_transform(df_proc[['R&D Spend']]),columns=['rnd0','rnd1','rnd2','rnd3'])
```

```python
ms=pd.DataFrame(pp.PolynomialFeatures(3).fit_transform(df_proc[['Marketing Spend']]),columns=['ms0','ms1','ms2','ms3'])
```

```python
df1=pd.concat([df_proc,rnd,ms],axis=1)
df1.head()
```

```python
df1.drop(['R&D Spend','Marketing Spend'],axis=1,inplace=True)
df1.head()
```

```python
x1=df1.drop('Profit',axis=1)
y1=df1['Profit']
```

```python
model1=LinearRegression().fit(x1,y1)
```

```python
model1.score(x1,y1)*100
```
