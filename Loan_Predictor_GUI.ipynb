```python
import pandas as pd
import tkinter as tk
import sklearn as skl
import warnings
warnings.filterwarnings('ignore')
```

```python
p = pd.read_pickle('loan_predictor.pkl')

app = tk.Tk()
app.geometry('450x400')
app.title('Loan Predictor')

##Background--->
#bg = tk.PhotoImage(file=)
#bg = bg.subsample(3)

canvas = tk.Canvas(app,width=450,height=300)
canvas.pack(fill='both', expand=True)
#canvas.create_image(0,0,image=bg, anchor='nw')
canvas.create_text(220,10,text='Loan Predictor')

def Predict():
    age = eval(Entry_age.get())
    exp = eval(Entry_exp.get())
    inc= eval(Entry_inc.get())
    fam = eval(Entry_fam.get())
    ccavg = eval(Entry_ccavg.get())
    edu = eval(Entry_edu.get())
    mort= eval(Entry_mort.get())
    secac= eval(Entry_secac.get())
    cd_ac = eval(Entry_cd_ac.get())
    online = eval(Entry_online.get())
    crec = eval(Entry_crec.get())

    quary = pd.DataFrame({'Age':[age],'Exp':[exp],'Inc':[inc],'Family':[fam],'CCAvg':[ccavg],'Education':[edu],'Mortage':[mort],'Sec_Acc':[secac],'CD_Acc':[cd_ac],'Onlie':[online],'CreditCard':[crec]})
    result.set(round(p.predict(quary)[0],2))

age = canvas.create_text(100,60,anchor='nw',text='Enter the Age')
exp = canvas.create_text(100,80,anchor='nw',text='Enter the Experience')
inc = canvas.create_text(100,100,anchor='nw',text='Enter the Income')
fam = canvas.create_text(100,120,anchor='nw',text='Enter the Family')
ccavg = canvas.create_text(100,140,anchor='nw',text='Enter the CCAvg')
edu = canvas.create_text(100,160,anchor='nw',text='Enter the Education')
mort = canvas.create_text(100,180,anchor='nw',text='Enter the Mortgage')
secac = canvas.create_text(100,200,anchor='nw',text='Enter the SEc._Acc.')
cd_ac = canvas.create_text(100,220,anchor='nw',text='Enter the CD_Acc.')
online = canvas.create_text(100,240,anchor='nw',text='Enter the Online')
crec = canvas.create_text(100,260,anchor='nw',text='Enter the CreditCard')

Entry_age = tk.Variable(app)
tk.Entry(app,textvariable=Entry_age).place(x=260,y=60,height=20)

Entry_exp = tk.Variable(app)
tk.Entry(app,textvariable=Entry_exp).place(x=260,y=80,height=20)

Entry_inc = tk.Variable(app)
tk.Entry(app,textvariable=Entry_inc).place(x=260,y=100,height=20)

Entry_fam = tk.Variable(app)
tk.Entry(app,textvariable=Entry_fam).place(x=260,y=120,height=20)

Entry_ccavg = tk.Variable(app)
tk.Entry(app,textvariable=Entry_ccavg).place(x=260,y=140,height=20)

Entry_edu = tk.Variable(app)
tk.Entry(app,textvariable=Entry_edu).place(x=260,y=160,height=20)

Entry_mort = tk.Variable(app)
tk.Entry(app,textvariable=Entry_mort).place(x=260,y=180,height=20)

Entry_secac = tk.Variable(app)
tk.Entry(app,textvariable=Entry_secac).place(x=260,y=200,height=20)

Entry_online = tk.Variable(app)
tk.Entry(app,textvariable=Entry_online).place(x=260,y=220,height=20)

Entry_cd_ac = tk.Variable(app)
tk.Entry(app,textvariable=Entry_cd_ac).place(x=260,y=240,height=20)

Entry_crec = tk.Variable(app)
tk.Entry(app,textvariable=Entry_crec).place(x=260,y=260,height=20)

tk.Button(app,text='Predictor',command=Predict,bg='light green').place(x=260,y=310,height=20)

Loan_Prediction = tk.Variable(app)
Loan_Prediction.set('Predict Loan')
tk.Label(app,textvariable=Loan_Prediction).place(x=260,y=330)

result = tk.Variable(app)
result.set('')
tk.Entry(app,textvariable=result).place(x=260,y=350,height=20)

app.mainloop()













```
