```python
import numpy as np
import pandas as pd
import tkinter as tk
import warnings
warnings.filterwarnings('ignore')
import sklearn as skl
```

```python
a = pd.read_pickle('Startup.pkl')

app=tk.Tk()
app.geometry('450x400')
app.title('Startup')

#bg=tk.Photoimage(file=)
#bg=bg.subsample(2)

canvas=tk.Canvas(app,width=450,height=300)
canvas.pack(fill='both',expand=True)
#canvas.create_image(0,0,image=bg,anchor='nw')

canvas.create_text(220,10,text='Startup Prediction')

def Predict():
    admin=eval(Entry_admin.get())
    california=eval(Entry_california_.get())
    florida=eval(Entry_florida_.get())
    new_york=eval(Entry_new_york_city.get())
    rnd0=eval(Entry_rn_0.get())
    rnd1=eval(Entry_rn_1.get())
    rnd2=eval(Entry_rn_2.get())
    rnd3=eval(Entry_rn_3.get())
    ms0=eval(Entry_ms_0.get())
    ms1=eval(Entry_ms_1.get())
    ms2=eval(Entry_ms_2.get())
    ms3=eval(Entry_ms_3.get())
    quary=pd.DataFrame({'Administration':[admin],'California':[california],'Florida':[florida],'NewYork':[new_york],'rnd0':[rnd0],'rnd1':[rnd1],'rnd2':[rnd2],'rnd3':[rnd3],'ms0':[ms0],'ms1':[ms1],'ms2':[ms2],'ms3':[ms3]})
    result.set(round(a.predict(quary)[0],2))

admin=canvas.create_text(100,60,anchor='nw',text='Enter the Admin Ammount')
california_=canvas.create_text(100,80,anchor='nw',text='Enter the Values(california)')
florida_=canvas.create_text(100,100,anchor='nw',text='Enter the Values(florida)')
new_york_city=canvas.create_text(100,120,anchor='nw',text='Enter the Values(new york)') 
rn_0=canvas.create_text(100,140,anchor='nw',text='Enter the rn0 Ammount') 
rn_1=canvas.create_text(100,160,anchor='nw',text='Enter the rn1 Ammount') 
rn_2=canvas.create_text(100,180,anchor='nw',text='Enter the rn2 Ammount') 
rn_3=canvas.create_text(100,200,anchor='nw',text='Enter the rn3 Ammount') 
ms_0=canvas.create_text(100,220,anchor='nw',text='Enter the ms0 Ammount') 
ms_1=canvas.create_text(100,240,anchor='nw',text='Enter the ms1 Ammount') 
ms_2=canvas.create_text(100,260,anchor='nw',text='Enter the ms2 Ammount') 
ms_3=canvas.create_text(100,280,anchor='nw',text='Enter the ms3 Ammount')

Entry_admin=tk.Variable(app)
tk.Entry(app,textvariable=Entry_admin).place(x=260,y=60,height=20)

Entry_california_=tk.Variable(app)
tk.Entry(app,textvariable=Entry_california_).place(x=260,y=80,height=20)

Entry_florida_=tk.Variable(app)
tk.Entry(app,textvariable=Entry_florida_).place(x=260,y=100,height=20)

Entry_new_york_city=tk.Variable(app)
tk.Entry(app,textvariable=Entry_new_york_city).place(x=260,y=120,height=20)

Entry_rn_0=tk.Variable(app)
tk.Entry(app,textvariable=Entry_rn_0).place(x=260,y=140,height=20)

Entry_rn_1=tk.Variable(app)
tk.Entry(app,textvariable=Entry_rn_1).place(x=260,y=160,height=20)

Entry_rn_2=tk.Variable(app)
tk.Entry(app,textvariable=Entry_rn_2).place(x=260,y=180,height=20)

Entry_rn_3=tk.Variable(app)
tk.Entry(app,textvariable=Entry_rn_3).place(x=260,y=200,height=20)

Entry_ms_0=tk.Variable(app)
tk.Entry(app,textvariable=Entry_ms_0).place(x=260,y=220,height=20)

Entry_ms_1=tk.Variable(app)
tk.Entry(app,textvariable=Entry_ms_1).place(x=260,y=240,height=20)

Entry_ms_2=tk.Variable(app)
tk.Entry(app,textvariable=Entry_ms_2).place(x=260,y=260,height=20)

Entry_ms_3=tk.Variable(app)
tk.Entry(app,textvariable=Entry_ms_3).place(x=260,y=280,height=20)

tk.Button(app,text='Predict Profit',command=Predict,bg='light green').place(x=260,y=310,height=20)

predict_profit=tk.Variable(app)
predict_profit.set('The Predicted Profit is \n In($)')
tk.Label(app,textvariable=predict_profit).place(x=260,y=330)

result=tk.Variable(app)
result.set('')
tk.Entry(app,textvariable=result).place(x=260,y=350,height=20)


app.mainloop()






```
